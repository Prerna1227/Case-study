```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType

# Define the schema based on the current DataFrame
schema = StructType([
    StructField("transaction_id", IntegerType(), True),
    StructField("product_id", IntegerType(), True),
    StructField("sales_amount", IntegerType(), True),
    StructField("sales_date", DateType(), True)
    # Add other fields based on the actual DataFrame schema
])

# Apply the schema to the DataFrame
df_new = df_new.selectExpr(
    "cast(transaction_id as int) as transaction_id",
    "cast(product_id as int) as product_id",
     "cast(sales_amount as int) as sales_amount",
    "cast(sales_date as date) as sales_date"
    # Add other fields based on the actual DataFrame schema
)

display(df_new)
```

```
from pyspark.sql.functions import col, year, month, sum

# Add a new column for the year of the sales_date
df_new = df_new.withColumn("sales_year", year(col("sales_date")))

# Add a new column for the month of the sales_date
df_new = df_new.withColumn("sales_month", month(col("sales_date")))

# Group by product_id and sales_year, and calculate the total sales_amount
df_new = df_new.groupBy("product_id", "sales_year").agg(sum("sales_amount").alias("total_sales_amount"))

display(df_new)
```
